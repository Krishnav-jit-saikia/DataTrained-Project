# 1. District wise crime commited IPC :

```python
df_1.head()
```

```python
df_1.columns
```

```python
# I will delete all the unimportant columns . 
unimportant_columns = ['CULPABLE HOMICIDE NOT AMOUNTING TO MURDER','CUSTODIAL RAPE','OTHER RAPE','KIDNAPPING AND ABDUCTION OF OTHERS','PREPARATION AND ASSEMBLY FOR DACOITY','AUTO THEFT','OTHER THEFT','RIOTS','CRIMINAL BREACH OF TRUST','COUNTERFIETING','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY','IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES']
df_1 = df_1.drop(unimportant_columns, axis = 1)
df_1
```

# Checking the dataset :

```python
df_1.isnull().sum()
```

```python
df_1.info()
```

```python
df_1.duplicated().sum()
```

# I will drop all the rows where the DISTRICT cotains TOTAL :

```python
Modified_df_1 = df_1[~df_1['DISTRICT'].str.contains('TOTAL')]
Modified_df_1
```

```python
# Now , fro  each district , I will  calculate the sum of each crime in for each year .
# By doing  this I will get for each district , how many times a specific crime happens from 2001 to 2012 .
# ( This process I will follow for each dataset .)
```

```python
exclude_cols = ['STATE/UT', 'DISTRICT', 'YEAR']

# Select crime columns (all columns except those to exclude)
crime_cols = [col for col in Modified_df_1.columns if col not in exclude_cols]
crime_cols
```

```python
total_IPC_crimes_by_district = Modified_df_1.groupby('DISTRICT')[crime_cols].sum()
total_IPC_crimes_by_district 
```

# Importing model :

```python
# I Thisnk KMeans Clstering is the best here 
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import  ColumnTransformer
from sklearn.model_selection import GridSearchCV
```

```python
columns_to_exclude = ['DISTRICT']
scaler = StandardScaler()
ohe = OneHotEncoder()
preprocessor = ColumnTransformer(
transformers = [
    ('onehotencoder', ohe ,[0] ),
    ('standardscaler', scaler ,[col for col in total_IPC_crimes_by_district.columns if col not in columns_to_exclude])
],
remainder = 'passthrough'
)
```

```python
kmeans = KMeans(n_clusters = 3 , init = 'k-means++', random_state = 0)

A = preprocessor.fit_transform(total_IPC_crimes_by_district)
y = kmeans.fit_predict(A)
y
```

```python
zero_indices = np.where(y == 0)[0]

# Print the indices
print(zero_indices)
```

```python
total_IPC_crimes_by_district.iloc[795]
```

```python
one_indices = np.where(y == 1)[0]

# Print the indices
print(one_indices)
```

```python
total_IPC_crimes_by_district.iloc[5]
```

```python
two_indices = np.where(y == 2)[0]

# Print the indices
print(two_indices)
```

```python
total_IPC_crimes_by_district.iloc[796]
```

```python
area_labels = {0: "Peaceful Area", 1: "Sensitive Area", 2: "Moderate Area"}

# Replace values in the array using the dictionary
data_replaced = np.vectorize(lambda x: area_labels.get(x))(y)
```

```python
data_replaced
```

```python
total_IPC_crimes_by_district['Area'] = data_replaced
```

```python
total_IPC_crimes_by_district
```

```python
# Convert the dataset into csv file :
filename = 'total_IPC_crimes_by_district.csv'

# Save the DataFrame to a CSV file
total_IPC_crimes_by_district.to_csv(filename, index=True)
```
